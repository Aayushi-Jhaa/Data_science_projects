```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cifar = tf.keras.datasets.cifar10
(x_train,y_train),(x_test,y_test) = cifar.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
i = 0
for i in range(10):
  plt.imshow(x_train[list(y_train).index(i)])
  plt.show(i+1)
+
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape =(32,32,3)))
model.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),activation="relu",padding="valid"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(tf.keras.layers.Flatten(input_shape=[32,32]))
model.add(tf.keras.layers.Dense(100,activation="relu"))
model.add(tf.keras.layers.Dense(10,activation="softmax"))
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
history = model.fit(x_train,y_train,batch_size=32,epochs=20,validation_split=0.1)
```

```python
#Training accuracy is very high,valkidation accuracy more than overfitting
model.evaluate(x_test,y_test)
```

```python
model.evaluate(x_train,y_train)
```

```python
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Conv2D(filters=5, kernel_size=(6, 6), activation='relu', input_shape=(32,32,3), ))
model1.add(tf.keras.layers.MaxPooling2D(pool_size=2,strides=2))
model1.add(tf.keras.layers.Conv2D(filters=5, kernel_size=(6,6), activation='relu'))
model1.add(tf.keras.layers.MaxPooling2D(pool_size=2,strides=2))
model1.add(tf.keras.layers.Flatten(input_shape=[32,32]))
model1.add(tf.keras.layers.Dense(units=120, activation='sigmoid'))
model1.add(tf.keras.layers.Dense(units=84, activation='sigmoid'))
model1.add(tf.keras.layers.Dense(units=10, activation = 'softmax'))
model1.summary()
```

```python
#model1.compile(loss=tcategorical_crossentropy, optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
model1.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
history = model1.fit(x_train,y_train,batch_size=32,epochs=20,validation_split=0.1)
```

```python
model1.evaluate(x_test,y_test)
```
